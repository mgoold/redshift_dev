```python
start_date = datetime.date(year=2023, month=1, day=1)
end_date = datetime.date(year=2023, month=1, day=3)

dt = fake.date_time_between(start_date=start_date, end_date=end_date)
sec_upperlimit=2000

sec_select1=[1,1,1,1,2,2,3,3,4,5,6,6,7,8,9] #low-tech way to get a distribution
											#sec_select1 picks a "sector" of the larger range
											#you're going to randomly pick from this list, which is weighted how you want
											#by overpopulating the values you want

sec_length=round(sec_upperlimit/len(sec_select1))
```

```python
rows=[]
dt = fake.date_time_between(start_date=start_date, end_date=end_date)

for i in range(0,sessioncount):
    userindx=userlist.index(userdist[i]) # goes back and picks individual userid out of set list
    uasitem=uas[userindx] # bc uas list only as long as unique user list, which make sense
    pagect=pagedist[i]

    for pg in range(0,pagect):
        randsecint=random.randint(0,len(sec_select1)-1) # interval from 2 sec to over 30 min
        sectoritem=sec_select1[randsecint]
        sector=sec_length*sectoritem

        extratime=random.randint(0,sec_length) 
        pagetime=sector+extratime

        dt=dt + datetime.timedelta(seconds=pagetime) # bump starting time by rand interval
        
        ln=[userdist[i],dt,uasitem]
        rows.append(ln)
        

```
